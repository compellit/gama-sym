## Stress-match on test-set

```python
import pandas as pd
```

```python
infi_test = "../../../out/scansion_gl_eval_202.annot.tsv"
```

```python
test_df = pd.read_csv(infi_test, sep="\t")
```

```python
test_df
```

```python
test_df.columns
```

```python
test_df['smt'] = test_df['smt'].astype(int)  # ensure numeric 0/1

result_test = (
    test_df.groupby('gsy')
      .agg(
          smt_ones=('smt', 'sum'),          # total 1s
          count=('smt', 'size')             # total rows in group
      )
)

# percentage of 1s within each gmt group
result_test['smt_pct'] = result_test['smt_ones'] / result_test['count'] * 100
```

```python
result_test
```

## Meter match on test-set

```python

test_df['mmt'] = test_df['mmt'].astype(int)  # ensure numeric 0/1

result_test_mmt = (
    test_df.groupby('gsy')
    .agg(
        mmt_ones=('mmt', 'sum'),  # total 1s
        count=('mmt', 'size')  # total rows in group
    )
)

# percentage of 1s within each gmt group
result_test_mmt['mmt_pct'] = result_test_mmt['mmt_ones'] / result_test_mmt['count'] * 100
```
